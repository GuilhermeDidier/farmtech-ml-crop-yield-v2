## 🎯 7. Detecção de Outliers

Vamos identificar outliers usando múltiplos métodos:
1. **IQR (Interquartile Range)**: Valores fora de [Q1 - 1.5×IQR, Q3 + 1.5×IQR]
2. **Z-Score**: Valores com |z-score| > 3
3. **Isolation Forest**: Algoritmo de detecção de anomalias

```python
# Função para detectar outliers usando IQR
def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detectar outliers em todas as variáveis numéricas
print('🔍 Detecção de Outliers usando Método IQR:')
print('='*90)

outliers_summary = []
for col in numeric_cols:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    outliers_summary.append({
        'Variável': col,
        'Qtd Outliers': len(outliers),
        'Percentual (%)': round(len(outliers) / len(df) * 100, 2),
        'Limite Inferior': round(lower, 2),
        'Limite Superior': round(upper, 2)
    })

outliers_df = pd.DataFrame(outliers_summary)
print(outliers_df.to_string(index=False))
```

```python
# Detecção de outliers usando Z-Score
print('\n🔍 Detecção de Outliers usando Z-Score (|z| > 3):')
print('='*90)

df_numeric = df[numeric_cols].copy()
z_scores = np.abs(stats.zscore(df_numeric))
outliers_zscore = (z_scores > 3).sum(axis=0)

zscore_summary = pd.DataFrame({
    'Variável': numeric_cols,
    'Outliers (Z-Score)': outliers_zscore,
    'Percentual (%)': (outliers_zscore / len(df) * 100).round(2)
})
print(zscore_summary.to_string(index=False))
```

```python
# Isolation Forest para detecção de outliers
print('\n🔍 Detecção de Outliers usando Isolation Forest:')
print('='*90)

# Preparar dados
X_outlier = df[numeric_cols].copy()

# Treinar Isolation Forest
iso_forest = IsolationForest(contamination=0.1, random_state=42, n_estimators=100)
outlier_predictions = iso_forest.fit_predict(X_outlier)

# -1 indica outlier, 1 indica normal
n_outliers_iso = (outlier_predictions == -1).sum()
print(f'Quantidade de outliers detectados: {n_outliers_iso}')
print(f'Percentual: {(n_outliers_iso/len(df)*100):.2f}%')

# Adicionar coluna de outliers ao dataframe
df['is_outlier'] = outlier_predictions == -1

print(f'\nOutliers por cultura:')
print(df.groupby('Crop')['is_outlier'].sum().sort_values(ascending=False))
```

## 📝 8. Conclusões da Análise Exploratória

### 📋 Principais Achados:

#### 1. Qualidade dos Dados
O dataset apresentou **excelente qualidade**: 156 registros sem valores nulos e sem duplicatas. Isso indica que os dados estão prontos para modelagem sem necessidade de imputação ou remoção de linhas inconsistentes.

#### 2. Composição do Dataset
O dataset é composto por **4 culturas agrícolas**, cada uma com exatamente **39 registros**, garantindo um balanceamento perfeito entre as classes:
- **Cocoa, beans** (Cacau)
- **Oil palm fruit** (Palma de óleo)
- **Rice, paddy** (Arroz)
- **Rubber, natural** (Borracha natural)

#### 3. Rendimento (Yield) — Alta Variabilidade Entre Culturas
A variável-alvo apresenta **altíssima variabilidade** (desvio padrão de 70.421 com média de 56.153), o que reflete diferenças fundamentais de escala entre as culturas:
- **Oil palm fruit** lidera com rendimento médio de ~175.804 kg/ha — valor muito superior às demais
- **Rice, paddy** apresenta rendimento médio de ~32.099 kg/ha
- **Rubber, natural** tem rendimento médio de ~10.571 kg/ha
- **Cocoa, beans** apresenta o menor rendimento médio (~8.883 kg/ha)

Essa diferença de escala é crítica e **explica diretamente por que modelos como SVR e MLP têm dificuldade** de generalizar sem normalização do target.

#### 4. Correlação das Variáveis com o Rendimento
As variáveis climáticas isoladas têm **correlação muito baixa** com o rendimento:
- Precipitação: 0.019
- Temperatura: 0.013
- Umidade específica: 0.013
- Umidade relativa: ~0.000

Isso indica que as condições climáticas são relativamente semelhantes entre todas as culturas, e que o **tipo de cultura (Crop) é o fator determinante** para o rendimento — o que foi confirmado pelos modelos preditivos.

#### 5. Detecção de Outliers
- Métodos **IQR** e **Z-Score** não identificaram outliers nas variáveis numéricas isoladas
- O **Isolation Forest** detectou **16 outliers (10.26%)**, concentrados principalmente em Oil palm fruit (12 casos), onde as variações de rendimento são naturalmente maiores

#### 6. Limitações Identificadas
- O dataset é pequeno (156 registros), o que pode limitar a capacidade de generalização dos modelos
- As variáveis climáticas têm baixa variabilidade entre si (precipitação entre 1.934 e 3.340 mm/dia), sugerindo que os dados foram coletados em uma região climática específica
- A ausência de variáveis de solo, fertilização e manejo agrícola limita a explicabilidade do modelo

#### 7. Conclusão Geral
O dataset está **limpo e bem estruturado**, mas apresenta um desafio importante: a diferença de escala entre culturas faz com que a variável alvo seja **multimodal** (com distribuições completamente diferentes por cultura). Modelos que não capturam bem essa estrutura — como regressão linear simples — tendem a ter baixo desempenho, enquanto modelos baseados em árvores (Random Forest, Gradient Boosting) se beneficiam naturalmente dessa separação.
